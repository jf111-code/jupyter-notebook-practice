## Why configuration matters

A Jupyter Notebook is more than just code cells.

Behind the notebook, there is:
- a Python environment
- a kernel that runs code
- settings and defaults that affect behavior

Understanding the setup helps prevent confusion and makes notebook work more reliable.

```python
import sys

print(sys.version)
```

## Python version

The output above shows the Python version currently being used by this notebook.

This matters because notebooks may behave differently depending on:
- Python version
- installed packages
- which kernel is selected

```python
import sys

print(sys.executable)
```

## Python executable

The path above shows which Python interpreter is running this notebook.

This helps answer an important question:

**Which Python environment am I actually using right now?**

```python
import os

print(os.getcwd())
```

## Working directory

The current working directory affects:
- where files are loaded from
- where files may be saved
- how relative paths behave

This is an important part of working with notebook projects cleanly.

```python
import platform

print("System:", platform.system())
print("Release:", platform.release())
print("Machine:", platform.machine())
```

## System information
