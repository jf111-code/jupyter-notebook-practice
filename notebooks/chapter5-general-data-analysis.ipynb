# General Data Analysis: Restaurant Tips Dataset

## Objective
Analyze tipping behavior to understand:
- What affects tip size?
- Are there patterns by time, gender, or group size?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
tips.describe()
```

```python
tips.columns
```

```python
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
tips.head()
```

```python
tips["tip_pct"].mean()
```

```python
tips.groupby("sex", observed=True)["tip_pct"].mean()
```

```python
tips.groupby("day", observed=True)["tip_pct"].mean()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Tip vs Total Bill")
plt.show()
```

```python
sns.histplot(tips["tip_pct"], bins=20)
plt.title("Distribution of Tip Percentage")
plt.show()
```

## Key Insights

- Tip percentage averages around X%
- Larger bills tend to generate higher absolute tips, but not always higher percentages
- Differences across days suggest behavioral patterns
