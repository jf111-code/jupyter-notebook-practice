## Why distribution matters

A notebook is more useful when it can be shared clearly.

Distribution matters because it helps other people:
- read the notebook
- understand the work
- reproduce the results
- see the project in a portfolio context

For me, distribution also matters because I want to publish notebook-based work on GitHub.

## Main ways notebooks can be shared

Based on Chapter 4, notebook distribution can include:

- GitHub
- Binder
- Google Colab
- running a notebook server

Each option supports a different kind of sharing experience.

## GitHub

GitHub is my main publishing platform for notebook work.

It helps me:
- store notebooks in repositories
- track versions with Git
- make notebooks visible as portfolio work
- organize projects with README files and links

At this stage, GitHub is the most important distribution method for me.

## Binder

Binder is useful when someone wants to launch a notebook in a live environment from a repository.

That is different from GitHub, which mainly renders the notebook for reading.

Binder seems useful for future projects where I want people to interact with notebooks instead of only viewing them.

## Google Colab

Google Colab is another way to run notebooks in the cloud.

It can be helpful for:
- quick experimentation
- notebook sharing
- cloud-based execution without local setup

For now, I want to focus first on building clean notebooks locally and publishing them to GitHub.

## What a shareable notebook needs

A notebook is easier to distribute when it has:

- a clear title
- short explanations
- readable section headings
- outputs that make sense
- a matching README in the repository
- a logical file location inside the project

```python
import os

notebook_folder = os.getcwd()
files = os.listdir(notebook_folder)

print("Current notebook folder:")
print(notebook_folder)
print("\nFiles in this folder:")
for file in files:
    print("-", file)
```

## Notebook organization

The code above shows the current working notebook folder and its files.

This is part of distribution too, because project structure affects how easy it is for someone else to navigate the repo.
